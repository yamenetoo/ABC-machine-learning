# **Tabular Data**

Tabular data refers to a structured form of data representation where information is organized into rows and columns, forming a table. Each row in the table represents a unique record or observation, and each column corresponds to a specific attribute or variable. Tabular data is commonly used to store and display information in a systematic and easily understandable format, making it suitable for various applications such as databases, spreadsheets, and datasets in data science. In the ***statistical analysis***, ***each row is considered a sample, representing an individual data point (one observation from sample), and each column represents a random variable***, capturing different aspects or features associated with the samples. This tabular structure facilitates efficient organization, analysis, and interpretation of data in a wide range of fields.

**Pandas** is one of the most popular and widely used Python libraries for working with ***tabular data***. It provides easy-to-use data structures like DataFrame, which is a two-dimensional table, as well as Series for one-dimensional data. Pandas is built on top of the NumPy library and is designed to work seamlessly with data from various sources, such as CSV files, Excel sheets, databases, and more.

Key features of Pandas for working with tabular data include:

1. **DataFrame (df):** A powerful data structure for working with tabular data, allowing easy indexing, slicing, and manipulation of data.

2. **Data Cleaning and Preprocessing:** Pandas provides functions for handling missing data, reshaping, merging, and aggregating data, making it a versatile tool for data cleaning and preprocessing.

3. **Data Analysis:** Pandas offers extensive functionality for data analysis, including groupby operations, statistical functions, and time-series analysis.

4. **Integration with Other Libraries:** Pandas integrates well with other popular Python libraries such as Matplotlib for plotting and visualization, and scikit-learn for machine learning tasks.

Due to its rich functionality and ease of use, Pandas has become a standard tool in the data science and analysis ecosystem, making it a go-to choice for working with tabular data in Python.

###Extract specific data from DataFrame

To extract specific columns from a pandas DataFrame, we can use the square bracket notation or the loc or iloc methods.

The following methods extracting specific columns from a pandas DataFrame.

1. **Using Column Name:**
    ```python
    extracted_col = df[column_name]
    ```
    Replace `column_name` with the name of the column you want to extract. This method is particularly useful when you know the names of the columns.

2. **Using Column Index:**
    ```python
    extracted_col = df.iloc[:, column_index]
    ```
    Replace `column_index` with the index of the column you want to extract. This method is handy when you want to extract columns based on their numerical position.

Both methods return a pandas Series containing the values of the specified column. If you want to extract multiple columns, you can pass a list of column names or indices.

Remember to adjust the `column_name` or `column_index` based on your actual DataFrame and the columns you want to extract.

````python
X1=df.iloc[:,0]                                                                 # ":" mean all row (samples) for spisific columns
X2=df.iloc[:,1]
X3=df.iloc[:,2]
X4=df.iloc[:,-2]
Y =df.iloc[:,-1]
````

The `df.iloc[row, columns]` syntax in pandas is used to index and select specific rows and columns based on their numerical positions.

- `row`: This specifies the index or indices of the row(s) you want to select.
- `columns`: This specifies the index or indices of the column(s) you want to select.

Here are a few examples:

1. Select a single element at the intersection of the 2nd row and 3rd column:
    ```python
    element = df.iloc[1, 2]
    ```

2. Select a specific row (2nd row) along with specific columns (3rd and 4th columns):
    ```python
    selected_data = df.iloc[1, 2:4]
    ```

3. Select multiple rows (2nd and 3rd rows) along with multiple columns (1st and 4th columns):
    ```python
    selected_data = df.iloc[1:3, [0, 3]]
    ```

4. Select all rows for a specific column (2nd column):
    ```python
    column_data = df.iloc[:, 1]
    ```

Remember that indices in Python are zero-based, so the first row or column has an index of 0. Adjust the row and column indices based on your specific DataFrame and the data you want to extract.

###Dropping row or columns

To drop rows or columns from a pandas DataFrame, you can use the `drop` method. Here are examples for dropping rows and columns:

1. **Dropping Rows:**
    ```python
    # Assuming 'index_to_drop' is the index of the row you want to drop
    df_dropped_rows = df.drop(index_to_drop)
    ```

    If you want to drop multiple rows, you can pass a list of indices:
    ```python
    # Assuming 'indices_to_drop' is a list of indices to drop
    df_dropped_rows = df.drop(indices_to_drop)
    ```

    Make sure to set `inplace=True` if you want the changes to be applied directly to the original DataFrame.

2. **Dropping Columns:**
    ```python
    # Assuming 'column_to_drop' is the name of the column you want to drop
    df_dropped_columns = df.drop(columns=column_to_drop)
    ```

    If you want to drop multiple columns, you can pass a list of column names:
    ```python
    # Assuming 'columns_to_drop' is a list of column names to drop
    df_dropped_columns = df.drop(columns=columns_to_drop)
    ```

    Similar to dropping rows, set `inplace=True` if you want the changes to be applied directly to the original DataFrame.

Always be careful when dropping rows or columns, and ensure that it aligns with your data processing needs. Adjust the variables (`index_to_drop`, `indices_to_drop`, `column_to_drop`, `columns_to_drop`) based on your specific use case and DataFrame.

````python
df_dropped_Y_column = df.drop(columns=["Y"])                                    #drop Y var from df
df_dropped_Y_column
````

If you want to modify the original DataFrame in-place, you can use the `inplace=True` parameter:

###Drop missing (NaN) values

To drop rows containing any NaN (Not a Number) values from a pandas DataFrame, we can use the `dropna()` method. Here's an example:


````python
import pandas as pd

My_df = pd.DataFrame({                                                          # Assuming My_df is DataFrame with NaN values
    'A': [1, 2, None, 4],
    'B': [5, 6, 7, 8],
    'C': [9, None, 11, 12]
})

My_df
````

````python
My_df_withOut_na = My_df.dropna()                                               # Drop rows with any NaN values
My_df_withOut_na
````

In this example, the dropna() method will remove any rows that contain at least one NaN value. The resulting DataFrame (df_no_na) will not include those rows.

If you want to drop columns with any NaN values instead, you can use df.dropna(axis=1).

Remember to adjust the code based on your specific DataFrame and requirements.

###Replace missing (NaN) values with the average of columns

To replace NaN values with the average of columns in a pandas DataFrame, you can use the `fillna` method along with the `mean` method. Here's an example:

````python
My_df
````

````python
My_df_filled = My_df.fillna(My_df.mean())                                          # Replace NaN values with the mean of each column
My_df_filled
````

#3.DataFrame descriptive summary

If you want to get a descriptive summary of a pandas DataFrame, you can use the `describe` method. Here's an example:


````python
My_df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 6, 7, 8, 9],
    'C': [10, 11, 12, 13, 14]
})

df_description = My_df.describe()


print(df_description)
````


The `describe` method provides various statistics, including count, mean, std (standard deviation), min, 25%, 50%, 75%, and max for each column in the DataFrame.

If your DataFrame contains non-numeric columns, `describe` will by default provide a summary of only the numeric columns. To include all columns, you can use `include='all'`:

````python
df_description_all = df.describe(include='all')
df_description_all
````

This will include count, unique, top, and freq for non-numeric columns.

Remember to adjust the code based on your specific DataFrame and requirements.

#4.plot a pandas DataFrame

To plot a pandas DataFrame, you can use the `plot` method, which is built on top of Matplotlib. The specific type of plot you use depends on the nature of your data and what kind of visualization you want. Here are a few examples:


**Line Plot:**

````python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 6, 7, 8, 9],
    'C': [10, 11, 12, 13, 14]
})

# Line plot
df.plot(x='A', y=['B', 'C'], kind='line')
plt.show()
````

**Bar Plot:**

````python
df = pd.DataFrame({
  'Category': ['A', 'B', 'C', 'D'],
  'Values': [10, 15, 7, 12]
})

# Bar plot
df.plot(x='Category', y='Values', kind='bar', rot=0)
plt.show()
````

**Histogram:**

````python
df = pd.DataFrame({
    'Values': [2, 4, 4, 6, 6, 6, 8, 8, 10]
})

# Histogram
df.plot(y='Values', kind='hist', bins=5, edgecolor='black')
plt.show()
````

These are just a few examples, and you can explore various other plot types and customization options available in pandas and Matplotlib based on your specific data and visualization needs. Adjust the columns, plot types, and parameters according to your DataFrame.

**scatter plot**

To create a scatter plot in pandas, you can use the plot method with kind='scatter'. Here's an example:


````python
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 6, 7, 8, 9],
    'C': [10, 11, 12, 13, 14]
})


df.plot(x='A', y='B', kind='scatter', title='A VS B Scatter Plot')              # Scatter plot
plt.show()
````

#5.Export a pandas DataFrame

To export a pandas DataFrame to a file, you can use various file formats such as CSV, Excel, JSON, or others, depending on your needs. Here are examples for exporting to CSV and Excel:


##**Export to CSV**

````python
import pandas as pd

# Assuming df is your DataFrame
# Example DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 6, 7, 8, 9],
    'C': [10, 11, 12, 13, 14]
})

# Export to CSV
df.to_csv('output.csv', index=False)
````

This will create a CSV file named `output.csv` in the current working directory.


##**Export to Excel**


````python
import pandas as pd

# Assuming df is your DataFrame
# Example DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 6, 7, 8, 9],
    'C': [10, 11, 12, 13, 14]
})

# Export to Excel
df.to_excel('output.xlsx', index=False)
````

This will create an Excel file named `output.xlsx` in the current working directory.

You can adjust the filename and path based on your preferences. The `index=False` parameter is used to exclude the index column from the exported file.

Also there are many type we can export to like **html**, **latex** and so on...
